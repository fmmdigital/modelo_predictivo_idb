```python
import pandas as pd
```

```python
df = pd.read_csv('prediccion_test_rf_409285.csv')
```

```python
df.head()
```

```python
df['Y_diferencia'] = df['Y_sobrecostos_suma'] != df['Y_pred']
```

```python
df.head()
```

```python
from sklearn import tree
X = df.loc[:, ~df.columns.isin(['Y_sobrecostos_suma', 'Y_pred','Y_diferencia'])].to_numpy()
Y = df['Y_diferencia'].to_numpy()
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
feat_importance = clf.tree_.compute_feature_importances(normalize=False)
```

```python
len(feat_importance)
```

```python
X
```

```python
importances = clf.tree_.compute_feature_importances(normalize=False)
important_features = pd.Series(data=importances,index=df.loc[:, ~df.columns.isin(['Y_sobrecostos_suma', 'Y_pred','Y_diferencia'])].columns)
important_features.sort_values(ascending=False,inplace=True)
```
